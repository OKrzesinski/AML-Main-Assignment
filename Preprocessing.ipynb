```python
#Indlæser data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=pd.read_csv('dataset_diabetes/diabetic_data.csv', na_values = '?')
df.head()
```

```python
# Print percent of missing values in each column
miss_val = df.isna().sum()/df.shape[0]*100
print(round(miss_val, 2))
```

```python
# identify patients that occur mulitple times and remove the rows after
df_sort = df.sort_values('patient_nbr', inplace = False)
df_dropped = df_sort.drop_duplicates(subset = 'patient_nbr', keep='first', inplace=False)

df_dropped.shape
```

```python
df['discharge_disposition_id'].dtypes
```

```python
#removing patients who have died or are expectec to die within 30 days (hospice)
discharget_to_remove = [11, 13, 14, 19, 20, 21]  #all the ways to die...
indexNames = []
for i in discharget_to_remove:
    curindex = []
    curindex = df_dropped[ df_dropped['discharge_disposition_id'] == i].index
    indexNames.extend(curindex) 
df_minus_dead = df_dropped.drop(indexNames, inplace=False)
print("indexNames: ", len(indexNames))
print(df_minus_dead.shape)
```

```python
#Dropping features with a lot of missing values
df_dropped_columns=df_minus_dead.drop(['weight','payer_code'], axis=1)
print(df_dropped_columns.shape)
```

```python
df_dropped_columns.to_csv(path_or_buf = 'dataset_diabetes/Reshaped_dataframe.csv', index = False)
```
